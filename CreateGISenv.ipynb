# Создание ГИС-среды Python
=================================

### Виртуальные среды Python
--------------------

Установка различных пакетов ГИС в Python иногда может быть немного сложной из-за различных зависимостей среди пакетов. Иногда для работы с пакетом может потребоваться более старая версия пакета или даже более старая версия Python. специфический инструмент для работы. Как уже упоминалось ранее, это также хорошая практика, чтобы установить как можно больше пакетов из одного и того же канал conda, и не смешивать conda и pip для установок, если это не строго необходимо. Самый простой способ заставить установку работать гладко-это построить специальный виртуальные реды Python для инструментов ГИС.

На практике среда python - это отдельная установка всех необходимых инструментов, включая интерпретатор Python, библиотеки и файлы сценариев. Python имеет собственную [virtualenv library](https://virtualenv.pypa.io/en/latest)
для создания изолированных сред Python. Во время этого курса мы рекомендуем использовать conda form для установки и управления Пакеты Python и мы познакомим вас с тем, как создавать
[Python environment с использованием conda](https://conda.io/docs/user-guide/tasks/manage-environments.html)
это включает в себя все инструменты, необходимые во время этого курса.

### Создайте среду с нуля

Conda имеет отличную документацию [о создании и управлении средами conda](https://docs.conda.io/projects/conda/en/latest/user-guide/tasks/manage-environments.html) где вы можете проверить детали используемых команд.

Основные этапы создания среды conda включают в себя:
    1) создание среды 
    2) активация среды 
    3) Установка пакетов

    1. Создайте среду и дайте ей имя:

    .. code-block::


        conda create --name python-gis


    2. Активируйте окружающую среду:

    .. code-block::

        conda activate python-gis

    Теперь вы должны увидеть имя среды в начале командной строки.

    2. Установите необходимые пакеты:

    .. code-block::

        # Установите jupyter лаборатории + ГИТ расширение
        conda install -c conda-forge jupyterlab
        conda install -c conda-forge jupyterlab-git

        # установить пакеты
        conda install -c conda-forge geopandas
        conda install -c conda-forge matplotlib
        conda install -c conda-forge geojson
        conda install -c conda-forge mapclassify
        conda install -c conda-forge contextily
        conda install -c conda-forge folium
        conda install -c conda-forge mplleaflet
        conda install -c conda-forge osmnx
        #  ... установите другие пакеты

    После установки всех необходимых пакетов можно приступить к работе в локальной лабораторной среде Jupyter, которая является связан с вашей средой python-gis conda, запустив Jupyter lab в командной строке.
    Cначала перейдитt к папке, где находятся ваши записные книжки Jupyter-файлы, прежде чем запускать Jupyter Lab.

    .. code-block::

        jupyter lab

    Note, Jupyter Lab will probably prompt you to "Build" the installation in order to get the git-plugin to show.

Создайте среду из файла YAML

Кроме того, можно создать среду conda на основе предварительно определенного конфигурационного файла. Требования к среде conda могут быть записаны в YAML-файл (расширение файла  .yaml или .yml).

Вы можете найти и скачать конфигурационный файл, содержащий все необходимые пакеты, необходимые во время курса autogis вот [здесь](https://github.com/Automating-GIS-processes/site/blob/master/ci/py38-GIS.yaml).

Содержимое конфигурационного файла выглядит следующим образом:

.. code-block::

    name: python-gis

    channels:
      - conda-forge
      - patrikhlobil

    dependencies:
      - python=3.8
      - jupyterlab
      - jupyterlab-git
      - matplotlib
      - geopandas
      - geojson
      - pysal
      - mapclassify
      - osmnx
      - pyrosm
      - geopy
      - geojson
      - rasterio
      - contextily
      - folium
      - mplleaflet
      - bokeh
      - patrikhlobil::pandas-bokeh
      - pip


После того как вы загрузили файл на свой собственный компьютер, вы можете перейти к этой папке и запустить эту команду для создания среда conda, использующая файл:

.. code-block::

    conda env create -f py38-GIS.yaml

Решение проблемы окружающей среды и установка всех пакетов может занять много времени, так что будьте терпеливы.

После завершения установки вы можете приступить к использованию пакетов ГИС, активировав среду.

.. code-block::

    source activate python-gis

Наконец, вы должны иметь возможность начать работать с Jupyter lab, активировав его в командной строке. Сначала перейдите к папке, где находятся ваши записные книжки Jupyter-файлы, прежде чем запускать Jupyter Lab.

.. code-block::

    jupyter lab


### Среды Docker
--------------------

Docker-это платформа, которая может быть использована для “упаковки” вычислительных инструментов в так называемый контейнер. Docker позволяет разрабатывать приложения и вычислительные среды, которые “готовы к запуску " без дальнейшие хлопоты с установками.

Например, экземпляры в записных книжках CSC основаны на образе docker, который содержит операционная система ubuntu с Jupyter Lab, Python и соответствующими пакетами Python для этого курса. Dockerfiles, используемые для настройки среды CSC Notebook для Geo-Python и AutoGIS, являются задокументировано на https://github.com/csc-training/geocomputing/.

Для целей этого курса мы рекомендуем студентам использовать файл YAML, для управления установками.
